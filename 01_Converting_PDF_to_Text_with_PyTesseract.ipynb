## Importing the PDF File and Converting the PDF to PNG Image Files
I am using the Pdf2Image library to create PNG images that can be feed into the PyTesseract Model.

```python
#  Store Pdf with convert_from_path function
images = convert_from_path('./data/notes_Redacted.pdf')
 
image_page_numbers = [] 
image_file_names = [] 
image_file_paths = []

for i in range(len(images)):
   
    # Save pages as images in the pdf
    image_file_name = 'page'+ str(i) +'.png'
    image_file_path = '/data/png_images/' + image_file_name
    images[i].save('.' + image_file_path, 'PNG')

    image_page_numbers.append(i)
    image_file_names.append(image_file_name)
    image_file_paths.append(image_file_path)
```

## Converting Images to Text with PyTesseract

```python
text_list = []

for file_path in image_file_paths:

  text = pytesseract.image_to_string(Image.open('.'+ file_path))
  text_list.append(text)
```

```python
data_dict = {
    "page_number":image_page_numbers, 
     "image_file_names": image_file_names, 
     "image_file_paths":image_file_paths,
     "text": text_list}

page_df = pd.DataFrame(data_dict)
```
