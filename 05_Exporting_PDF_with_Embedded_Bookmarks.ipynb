## Exporting PDF with Embedded Bookmarks
The data extracted with Regular Expressions for the Lenders variable yielded the highest level of accuracy (over the NER Model for the Lenders variable), and will be used for my final submission.

The ML Classifier actually yielded 100% accuracy, but given the small sample size of the test dataset(15 documents), this model needs to be validated on a larger test set. For the purposes of this submission I will use the Document Label variables extracted with Regular Expressions.

We will utilize the PyPDF2 Library to add the embedded bookmarks the original PDF file.

```python
results = pd.read_csv('./data/coding_challenge_results_extracted_data.csv', index_col=0)
```

```python
results.head()
```

```python
label_list = []

for idx,i in enumerate(results['document_label']):
  label = f'{i}_note_{idx}'
  label_list.append(label)

label_list[:5]  
```

```python
writer = PdfFileWriter() 
reader = PdfFileReader(open('./data/notes_Redacted.pdf', 'rb')) # open input

for i in range(len(label_list)):
  writer.addPage(reader.getPage(i)) 
  writer.addBookmark(title=label_list[i], pagenum=i)

output = open('./data/PDF_with_embedded_booksmarks.pdf','wb') 
writer.write(output) 
output.close() 
```
